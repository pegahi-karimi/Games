```python
values = {'Ace':11,'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':11,'Queen':12,'King':13}
suits = ['Hearts','Diamonds','Spades','Clubs']
ranks = ['Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King']
```

```python
import random
```

```python
class Account:
    
    def __init__(self,balance):
        self.balance = balance
        
    def bet(self,bet_amt):          
        while self.balance < int(bet_amt):
            print('Bet not accepted.')
            bet_amt = input('Enter bet:')
            
        self.bet_amt = int(bet_amt)
        print('Bet accepted!')
        
    def win(self):
        self.balance += self.bet_amt
        
    def lose(self):
        self.balance -= self.bet_amt
    
    def __str__(self):
        return f'The current balance is {self.balance}'
        
```

```python
class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
                
    def __str__(self):
        return f'{self.rank} of {self.suit} with a value of {self.value}'
        
```

```python
class Deck:
    
    def __init__(self):   
        self.all_cards = []
        
        for rank in ranks:
            for suit in suits:
                self.all_cards.append(Card(suit,rank))
        
    def shuffle(self):
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        return self.all_cards.pop()
    
```

```python
class Player:
    
    def __init__(self):
        self.hand = []
        
```

```python
def Ace_check(a,l):
    if a>21:
        for card in l:
            if card.rank == 'Ace' and a>21:
                l.remove(card)
                a-=10
    return a,l
```

```python
deck = Deck()

deck.shuffle()

player = Player()

comp = Player()

player_account = Account(int(input('What is your balance?')))


game_on = True
m = 0
c = 0

while game_on:
    
    player_bet = player_account.bet(int(input('How much do you want to bet?')))

    #At the beginning of the game
    
    if len(player.hand) == 0:
        print("\nPlayer's hand is:")
        player.hand.append(deck.deal_one())
        print(player.hand[0])
        
        player.hand.append(deck.deal_one())
        print(player.hand[1])
        
        comp.hand.append(deck.deal_one())
        comp.hand.append(deck.deal_one())
    
    resp = 'Y'
    
    #When player wants to continue
    
    while resp == 'Y':

        if m == 0:
            for card in player.hand:
                m+=card.value
            print(f'\nThe total value of cards for player is {m}.')
        
        while m < 21:
            resp = input('Another card? Y or N ').upper()
            if resp == 'Y':
                player.hand.append(deck.deal_one())
                print(player.hand[-1])
                m+=player.hand[-1].value
                print(f'\nThe total value of cards for player is {m}.')
            else:
                resp = 'N'
                break
        
        if m == 21:
            break
            
        elif m > 21:
            m,player.hand = Ace_check(m,player.hand)
            if m < 21:
                resp = 'Y'
            else:
                resp = 'N'
                
    print(f'Player is equal to {m}.\n')
    
    #Computer's turn if player didn't lose
    
    if m <= 21:
        print("Now it is computer's turn.\n\nComputer's hand is:")

        if c == 0:
            for card in comp.hand:
                c+=card.value
                print(card)

        comp_on = True
        
        while comp_on:
            while c<21 and c<m:
                comp.hand.append(deck.deal_one())
                c+=comp.hand[-1].value
                print(comp.hand[-1])

            if c>21:
                c,comp.hand = Ace_check(c,comp.hand)
                if c<21:
                    comp_on = True
                else:
                    comp_on = False
            else:
                comp_on = False
        
        print(f'Comp is equal to {c}.')
        
        if c<=21:
            print('\nYou lost by having lower cards!\n')
            win = False
            break
        else:
            print('\nYou won!\n')
            win = True
            break
            
    else:
        print('You lost by reaching 21!\n')
        win = False
        game_on = False
        
    
        
    if win == True:
        player_account.win()
    else:
        player_account.lose()

    print(player_account)
    
    if player_account.balance > 0:
        y = input('Do you want to play again? Y or N ')
        if y.upper() == 'Y':
            game_on = True
```
